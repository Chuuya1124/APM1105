I. amazon.csv

About the Dataset
Context
Forest fires are a serious problem for the preservation of the Tropical Forests. Understanding the frequency of forest fires in a time series can help to take action to prevent them.

Brazil has the largest rainforest on the planet that is the Amazon rainforest.

Content
This dataset reports the number of forest fires in Brazil divided by states. The series comprises the period of approximately 10 years (1998 to 2017). The data were obtained from the official website of the Brazilian government.

http://dados.gov.br/dataset/sistema-nacional-de-informacoes-florestais-snifLinks to an external site.

For data exploration and visualization, provide each of the following items:

1. Display top 10 rows of the dataset.
2. Check last 10 rows.
3. Find shape of the dataset (number of rows and number of columns).
4. Get information about the dataset like total number rows, total number of columns, datatypes of each column and memory requirement.
5. Check for duplicate data and drop them.
6. Check null values in the dataset.
7. Get overall statistics about the dataframe.
8. Rename month names to English.
9. Get the total number of fires registered.
10. In which month maximum number of forest fires were reported?
11. In which year maximum number of forest fires was reported?
12. In which state maximum number of forest fires was reported?
13. Find total number of fires were reported in Amazonas.
14. Display number of fires were reported in Amazonas (year-wise).
15. Display number of fires were reported in Amazonas (day-wise).
16. Find total number of fires were reported in 2015 and visualize data based on each ‘month’.
17. Find average number of fires were reported from highest to lowest (state-wise).
18. Find the state names where fires were reported in 'dec' month.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("amazon.xlsx")

#1. Top 10 rows of the dataset
print("Top 10 rows of the dataset:")
print(df.head(10))
print("\n")

#2. Last 10 rows
print("Last 10 rows of the dataset:")
print(df.tail(10))
print("\n")

#3. Find shape of the dataset (number of rows and number of columns)
print("Shape of the dataset (rows, columns):", df.shape)
print("\n")

#4. Information about data set
print("Information about the dataset:")
print(df.info())
print("\n")

#5. Check for duplicate data and drop them
df.drop_duplicates(inplace=True)

#6. Check null values in the dataset
print("Null values in the dataset:")
print(df.isnull().sum())
print("\n")

#7. Overall statistics about the dataframe
print("Overall statistics about the dataframe:")
print(df.describe())
print("\n")

#8. Rename month names to English
month_mapping = {'Janeiro': 'January', 'Fevereiro': 'February', 'Março': 'March', 'Abril': 'April', 'Maio': 'May', 'Junho': 'June', 'Julho': 'July', 'Agosto': 'August', 'Setembro': 'September', 'Outubro': 'October', 'Novembro': 'November', 'Dezembro': 'December'}
df['month'] = df['month'].map(month_mapping)

#Total number of fires registered
total_fires = df['number'].sum()
print("Total number of fires registered:", total_fires)
print("\n")

#10. In which month maximum number of forest fires were reported?
max_fires_month = df.loc[df['number'].idxmax()]['month']
print("Month with maximum number of forest fires reported:", max_fires_month)
print("\n")

#11. In which year maximum number of forest fires was reported?
max_fires_year = df.loc[df['number'].idxmax()]['year']
print("Year with maximum number of forest fires reported:", max_fires_year)
print("\n")

#12. In which state maximum number of forest fires was reported?
max_fires_state = df.loc[df['number'].idxmax()]['state']
print("State with maximum number of forest fires reported:", max_fires_state)
print("\n")

#13. Find total number of fires were reported in Amazonas
total_fires_amazonas = df[df['state'] == 'Amazonas']['number'].sum()
print("Total number of fires reported in Amazonas:", total_fires_amazonas)
print("\n")

#14. Display number of fires were reported in Amazonas (year-wise)
print("Number of fires reported in Amazonas (year-wise):")
print(df[df['state'] == 'Amazonas'].groupby('year')['number'].sum())
print("\n")

# Plot number of fires reported in Amazonas (year-wise)
plt.figure(figsize=(12, 6))
amazonas_yearly_fires = df[df['state'] == 'Amazonas'].groupby('year')['number'].sum()
plt.plot(amazonas_yearly_fires, marker='o', linestyle='-')
plt.title("Number of fires reported in Amazonas (year-wise)")
plt.xlabel("Year")
plt.ylabel("Number of Fires")
plt.grid(True)
plt.show()

# Check if 'date' column exists before displaying date-wise data
if 'date' in df.columns:
    # Display number of fires were reported in Amazonas (date-wise)
    print("Number of fires reported in Amazonas (date-wise):")
    print(df[df['state'] == 'Amazonas'].groupby('date')['number'].sum())
    print("\n")
else:
    print("'date' column does not exist in the dataset.")
    print("\n")

# Find total number of fires were reported in 2015 and visualize data based on each ‘month’
fires_2015 = df[df['year'] == 2015]
print("Total number of fires reported in 2015:", fires_2015['number'].sum())
print("\n")

# Plot number of fires reported in 2015 (month-wise)
plt.figure(figsize=(10, 6))
fires_2015_monthly = fires_2015.groupby('month')['number'].sum()
fires_2015_monthly.plot(kind='bar', color='skyblue')
plt.title("Number of fires reported in 2015 (month-wise)")
plt.xlabel("Month")
plt.ylabel("Number of Fires")
plt.xticks(rotation=45)
plt.show()

# Find average number of fires were reported from highest to lowest (state-wise)
print("Average number of fires reported (state-wise):")
average_fires_state = df.groupby('state')['number'].mean().sort_values(ascending=False)
print(average_fires_state)
print("\n")

# Plot average number of fires reported (state-wise)
plt.figure(figsize=(12, 8))
average_fires_state.plot(kind='bar', color='orange')
plt.title("Average number of fires reported (state-wise)")
plt.xlabel("State")
plt.ylabel("Average Number of Fires")
plt.xticks(rotation=45)
plt.show()

# Find the state names where fires were reported in 'dec' month
states_dec = df[df['month'] == 'December']['state'].unique()
print("State names where fires were reported in 'dec' month:", states_dec)
```

II. MovieLens 25M Dataset

Questions:
Note: Aside from tables, display charts/graphs for the items below.

1. How many movies are released each year? What is the percentage of change over time?
2. What are the most popular genre of movies?
3. Does the taste of movies changes over time?
4. What are the top 20 movies that has the top amount of users rating and has the highest rate?
5. What is the most of popular season for releasing movies?

```python
import pandas as pd
import matplotlib.pyplot as plt

ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
genome_scores = pd.read_csv("genome-scores.csv")
genome_tags = pd.read_csv("genome-tags.csv")
links = pd.read_csv("links.csv")
tags = pd.read_csv("tags.csv")

# Merge (ratings and movies)
ratings_movies = pd.merge(ratings, movies, on='movieId')

# timestamp to datetime format
ratings_movies['timestamp'] = pd.to_datetime(ratings_movies['timestamp'], unit='s')

# Extract year(timestamp)
ratings_movies['year'] = ratings_movies['timestamp'].dt.year

#1. How many movies are released each year? What is the percentage of change over time?
movies_per_year = ratings_movies.groupby('year')['movieId'].nunique()
percentage_change = movies_per_year.pct_change()

#number of movies released each year
plt.figure(figsize=(12, 6))
plt.bar(movies_per_year.index, movies_per_year, color='skyblue')
plt.title("Number of Movies Released Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

#percentage change over time
plt.figure(figsize=(12, 6))
plt.plot(percentage_change.index, percentage_change, marker='o', color='orange')
plt.title("Percentage Change of Movies Released Over Time")
plt.xlabel("Year")
plt.ylabel("Percentage Change")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

#2. Most popular genres of movies?
genres = movies['genres'].str.split('|', expand=True).stack().value_counts().head(10)

# Plot most popular genres
plt.figure(figsize=(10, 6))
genres.plot(kind='bar', color='green')
plt.title("Most Popular Genres of Movies")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

#3. Does the taste of movies change over time?
ratings_per_year = ratings_movies.groupby('year')['rating'].mean()

# Plot average rating over time
plt.figure(figsize=(12, 6))
plt.plot(ratings_per_year.index, ratings_per_year, marker='o', color='red')
plt.title("Average Rating of Movies Over Time")
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

#4. What are the top 20 movies that have the most user ratings and highest rating?
top_movies = ratings_movies.groupby('title').agg({'rating': ['count', 'mean']})
top_movies.columns = ['num_ratings', 'avg_rating']
top_movies = top_movies.sort_values(by=['num_ratings', 'avg_rating'], ascending=[False, False]).head(20)

# Plot top 20 movies with most user ratings and highest rating
plt.figure(figsize=(12, 8))
plt.barh(top_movies.index[::-1], top_movies['num_ratings'].values[::-1], color='blue', alpha=0.7)
plt.xlabel("Number of Ratings")
plt.ylabel("Movie Title")
plt.title("Top 20 Movies with Most User Ratings and Highest Rating")
plt.tight_layout()
plt.show()

#5. What is the most popular season for releasing movies?
# Extract month from timestamp
ratings_movies['month'] = ratings_movies['timestamp'].dt.month

#Mapping for seasons
seasons_mapping = {1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring', 6: 'Summer',
                   7: 'Summer', 8: 'Summer', 9: 'Fall', 10: 'Fall', 11: 'Fall', 12: 'Winter'}

# (months to seasons)
ratings_movies['season'] = ratings_movies['month'].map(seasons_mapping)

movies_per_season = ratings_movies.groupby('season')['movieId'].nunique()

# Plot number of movies released in each season
plt.figure(figsize=(10, 6))
movies_per_season.plot(kind='bar', color='purple')
plt.title("Number of Movies Released in Each Season")
plt.xlabel("Season")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
