```python

import requests
import random

# URL of the raw dataset file on Github
url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'

# Send a GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content of the response to a file
    with open('words.txt', 'wb') as f:
        f.write(response.content)
    print('Dataset saved to file.')
else:
    print('Failed to download dataset.')

def count_total_words():
    with open('words.txt', 'r') as file:
        words = file.readlines()
    return len(words)

def find_longest_word():
    with open('words.txt', 'r') as file:
        words = file.readlines()
    longest_word = max(words, key=len).strip()
    return longest_word

def count_words_start_with(letter):
    with open('words.txt', 'r') as file:
        words = file.readlines()
    count = sum(1 for word in words if word.startswith(letter))
    return count

def print_words_contain_substring(substring):
    with open('words.txt', 'r') as file:
        words = file.readlines()
    words_contain_substring = [word.strip() for word in words if substring in word]
    return words_contain_substring

def print_palindromes():
    with open('words.txt', 'r') as file:
        words = file.readlines()
    palindromes = [word.strip() for word in words if word.strip() == word.strip()[::-1]]
    return palindromes

def count_letter_frequency():
    frequency = {}
    with open('words.txt', 'r') as file:
        words = file.read().replace('\n', '')
    for letter in words:
        frequency[letter] = frequency.get(letter, 0) + 1
    return frequency

def word_guessing_game():
    def select_word():
        with open('words.txt', 'r') as file:
            words = file.readlines()
        return random.choice(words).strip()

    def display_word(word, guessed_letters):
        display = ''
        for letter in word:
            if letter in guessed_letters:
                display += letter
            else:
                display += '_'
        return display

    def get_guess():
        guess = input("Guess a letter: ").lower()
        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter.")
            return get_guess()
        return guess

    word_to_guess = select_word()
    guessed_letters = set()
    remaining_attempts = 8

    print("Welcome to the Word Guessing Game!")
    print("Try to guess the word.")

    while remaining_attempts > 0:
        print("\nWord:", display_word(word_to_guess, guessed_letters))
        print("Remaining attempts:", remaining_attempts)
        guess = get_guess()
        guessed_letters.add(guess)
        remaining_attempts -= 1
        if guess in word_to_guess:
            print("Correct!")
        else:
            print("Incorrect!")
        if all(letter in guessed_letters for letter in word_to_guess):
            print("\nGood Job! You guessed the word:", word_to_guess)
            break
    else:
        print("\nSorry, you lose. The word was:", word_to_guess)

# Print the total number of words in the file
print("Total number of words in the file:", count_total_words())

# Print the longest word in the file
print("Longest word in the file:", find_longest_word())

# Print the number of words that start with a certain letter (use user input)
letter = input("Enter a letter: ").lower()
print("Number of words that start with '{}':".format(letter), count_words_start_with(letter))

# Print all the words that contain a certain substring (use user input)
substring = input("Enter a substring: ").lower()
print("Words that contain '{}':".format(substring), print_words_contain_substring(substring))

# Print all the words that are palindromes
print("Palindromes:", print_palindromes())

# Count the frequency of each letter
print("Letter frequency:", count_letter_frequency())

# Run the word guessing game
word_guessing_game()
```
